###### What is clustering in machine learning?
***

"Clustering is a Machine Learning technique that involves the grouping of data points. Given a set of data points, we can use a clustering algorithm to classify each data point into a specific group. In theory, data points that are in the same group should have similar properties and/or features, while data points in different groups should have highly dissimilar properties and/or features. Clustering is a method of unsupervised learning and is a common technique for statistical data analysis used in many fields." [7]

###### What is k-means clustering?
***

According to the "Oracle AI and Data Science Blog" written by Andrea Trevino k-means clustering is "is a type of unsupervised learning, which is used when you have unlabeled data (i.e., data without defined categories or groups). The goal of this algorithm is to find groups in the data, with the number of groups represented by the variable K. The algorithm works iteratively to assign each data point to one of K groups based on the features that are provided. Data points are clustered based on feature similarity."[5]

###### Why use k-means clustering?
***

K-means clustering comes with a variety of reasons why to use it. Firstly its simplicity. The majority of the K-means algorithm  is carried out in 2 steps. Cluster assignment and move centroid. So If we’re looking for an unsupervised learning algorithm that is easy to implement and can handle large data sets, K-means clustering is a good starting point. 

Secondly it is so widely available. The majority of machine learning packages will contain an implementation of the k-means clustering algorithm.

Lastly the speed of the algorithm. This algorithm is very fast and works quickly even with large datasets. This shows that using the k-means algorithm offers great advantages over others and for all of the reasons listed above you can get a good idea of how the algorithm is good.



###### Shortfalls of k-means clustering
***

Although k-means clustering has many advantages like anything it has some shortfalls too as nothing is perfect! Firstly k-means clustering doesn't allow data points that are far away from eachother share the same cluster even though they belong in the same cluster.

Another set back of k-means clustering is the algorithm assumes spherical shapes of clusters and doesn’t work well when clusters are in different shapes.

K-means clustering also requires prior specification for the number of cluster centers.

Lastly If there are two highly overlapping data then it cannot be distinguished and cannot tell that there are two clusters.

###### The Code
***

```python
# Imports
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans

# import warnings filter - reference [1]
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)

#Loading the dataset
iris = datasets.load_iris()

#Defining targets and predictors
X = iris.data[:, :2]
y = iris.target
```

```python
#Scatter Plot
plt.scatter(X[:,0], X[:,1], c=y, cmap='brg')
plt.xlabel('Length', fontsize=18)
plt.ylabel('Width', fontsize=18)
```

```python
# Cluster model using 3 clusters
km = KMeans(n_clusters = 3, n_jobs = 4, random_state=21)
km.fit(X)
```

```python
#Identifying the centre points
centers = km.cluster_centers_
print("The centre points")
print()
print(centers)
```

```python
#this will tell us to which cluster does the data observations belong.
new_labels = km.labels_
# Plot the identified clusters and compare with the answers
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='brg',
edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='cubehelix',
edgecolor='k', s=150)
axes[0].set_xlabel('Length', fontsize=15)
axes[0].set_ylabel('Width', fontsize=15)
axes[1].set_xlabel('Length', fontsize=15)
axes[1].set_ylabel('Width', fontsize=15)
axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[0].set_title('REAL', fontsize=15)
axes[1].set_title('PREDICTED', fontsize=15)
```
